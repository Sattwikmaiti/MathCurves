```python
import numpy as np
import matplotlib.pyplot as plt

x=range(-800,800,1);
y=[]

for i in range(len(x)):
  
  if(x[i]<0):
     temp=0
  else: 
     temp=1
  y.append(temp)   

#plotting the graph
plt.plot(x,y)
plt.xlabel('t -->')
plt.ylabel('u(t) -->')
plt.title('unit step function')
plt.grid()
plt.show()
```
