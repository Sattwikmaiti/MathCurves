```python
import numpy as np
import matplotlib.pyplot as plt

x1=range(-50,50,1);
y1=[]

for i in range(len(x1)):
  if(x1[i]>=0 and x1[i]<5):
     temp=1
  elif(x1[i]>=5 and x1[i]<8):
     temp=2
  elif(x1[i]>=8 and x1[i]<12):
     temp=5
  else:
    temp=0        
  
  y1.append(temp)   

#plotting the graph
plt.plot(x1,y1)
plt.xlabel('t -->')
plt.ylabel('u(t) -->')
plt.title('X(t)')
plt.grid()
plt.show()

x2=range(-50,50,1);
y2=[]

for i in range(len(x2)):
  if(x2[i]>=0 and x2[i]<7):
     temp=2
  elif(x2[i]>=7 and x2[i]<10):
     temp=0
  elif(x2[i]>=10 and x2[i]<15):
     temp=7
  else:
    temp=0        
  
  y2.append(temp)   

#plotting the graph
plt.plot(x2,y2)
plt.xlabel('t -->')
plt.ylabel('u(t) -->')
plt.title('Y(t)')
plt.grid()
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt

def Y(t):
  if(t>=0 and t<7):
     temp=2
  elif(t>=7 and t<10):
     temp=0
  elif(t>=10 and t<15):
     temp=7
  else:
    temp=0  
  return temp;   
def X(t):
  if(t>=0 and t<5):
     temp=1
  elif(t>=5 and t<8):
     temp=2
  elif(t>=8 and t<12):
     temp=5
  else:
    temp=0 
  return temp;   
t=range(-50,50,1)
ans=[]

for i in range(len(t)):
  ans.append(X(t[i])+Y(t[i]))

  

#plotting the graph
plt.plot(t,ans)
plt.xlabel('t -->')
plt.ylabel('u(t) -->')
plt.title('x(t)+y(t)')
plt.grid()
plt.show()

t=range(-50,50,1)
ans=[]

for i in range(len(t)):
  ans.append(X(t[i])-Y(t[i]))

  

#plotting the graph
plt.plot(t,ans)
plt.xlabel('t -->')
plt.ylabel('u(t) -->')
plt.title('x(t)-y(t)')
plt.grid()
plt.show()

t=range(-50,50,1)
ans=[]

for i in range(len(t)):
  ans.append(X(t[i])*Y(t[i]))

  

#plotting the graph
plt.plot(t,ans)
plt.xlabel('t -->')
plt.ylabel('u(t) -->')
plt.title('x(t)*y(t)')
plt.grid()
plt.show()
t=range(-50,50,1)
ans=[]

for i in range(len(t)):
  ans.append((X(t[i])/2)+(Y(t[i]/3)))

  

#plotting the graph
plt.plot(t,ans)
plt.xlabel('t -->')
plt.ylabel('u(t) -->')
plt.title('x(t)/2 + y(t)/3')
plt.grid()
plt.show()


t=range(-50,50,1)
ans=[]

for i in range(len(t)):
  ans.append(X(-t[i]))

  

#plotting the graph
plt.plot(t,ans)
plt.xlabel('t -->')
plt.ylabel('u(t) -->')
plt.title('-X(t)')
plt.grid()
plt.show()


t=range(-50,50,1)
ans=[]

for i in range(len(t)):
  ans.append(Y(-t[i]))

  

#plotting the graph
plt.plot(t,ans)
plt.xlabel('t -->')
plt.ylabel('u(t) -->')
plt.title('-Y(t)')
plt.grid()
plt.show()

t=range(-50,50,1)
ans=[]

for i in range(len(t)):
  ans.append(X(2*t[i]))

  

#plotting the graph
plt.plot(t,ans)
plt.xlabel('t -->')
plt.ylabel('u(t) -->')
plt.title('X(2t)')
plt.grid()
plt.show()


t=range(-50,50,1)
ans=[]

for i in range(len(t)):
  ans.append(X(-2*t[i]+5))

  

#plotting the graph
plt.plot(t,ans)
plt.xlabel('t -->')
plt.ylabel('u(t) -->')
plt.title('X(-2t+5)')
plt.grid()
plt.show()


t=range(-50,50,1)
ans=[]

for i in range(len(t)):
  ans.append(X(0.5*t[i]-5))

  

#plotting the graph
plt.plot(t,ans)
plt.xlabel('t -->')
plt.ylabel('u(t) -->')
plt.title('X(0.5t-5)')
plt.grid()
plt.show()


t=range(-50,50,1)
ans=[]

for i in range(len(t)):
  ans.append(X(-0.5*t[i]-5))

  

#plotting the graph
plt.plot(t,ans)
plt.xlabel('t -->')
plt.ylabel('u(t) -->')
plt.title('X(-0.5t-5)')
plt.grid()
plt.show()



```
