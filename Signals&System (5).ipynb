```python
import numpy as np
import matplotlib.pyplot as plt

x=range(-800,800,1);
y=[]

for i in range(len(x)):
  
  if(x[i]<0):
     temp=0
  else: 
     temp=1
  y.append(temp)   

#plotting the graph
plt.plot(x,y)
plt.xlabel('t -->')
plt.ylabel('u(t) -->')
plt.title('unit step function')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
x=range(-800,800,1);
y=[]

for i in range(len(x)):
  
  temp =math.sin(math.radians(x[i]))
  y.append(temp)   

#plotting the graph
plt.plot(x,y)
plt.xlabel('t -->')
plt.ylabel('u(t) -->')
plt.title('sine function')
plt.grid()
plt.show()
```

```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
#fc: Center frequency.
#retquad: If True, return the imaginary as well as the real part of the signal. Default is False.
#retenv: If True, return the envelope of the signal. Default is False
#i: The real part of the signal. Always returned.
#q:  Imaginary part of signal. Only returned if retquad is True.
#e:  The envelope of the signal. Only returned if retenv is True.
t = np.linspace(-1, 1, 200)

i, q, e = signal.gausspulse(t, fc=3, retquad=True, retenv=True)

plt.plot(i, color='green')
plt.plot(q, color='black')
plt.plot(e, '--', color='red')

plt.title('Gauss pulse for a 3 Hz ')
plt.ylabel("Normalized magnitude [dB]")
plt.xlabel("Normalized frequency [cycles per sample]")
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.legend(['Inphase', 'Quadrature', 'Envelope'])

plt.show()
```
